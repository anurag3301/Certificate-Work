#### Lets make consequent group of 20 countries and analyse there scores in each field in a clean graph.

```python
data = {
    "Groups":[],
    "GDP per capita":[],
    "Social support":[],
    "Healthy life expectancy":[],
    "Freedom to make life choices":[],
    "Generosity":[],
    "Perceptions of corruption":[]
}
```

```python
counter = 0
for i in range(1, 9):
    gdp = data_2019.loc[counter:counter+20]["GDP per capita"].mean()
    social = data_2019.loc[counter:counter+20]["Social support"].mean()
    helth = data_2019.loc[counter:counter+20]["Healthy life expectancy"].mean()
    freedom = data_2019.loc[counter:counter+20]["Freedom to make life choices"].mean()
    generosity = data_2019.loc[counter:counter+20]["Generosity"].mean()
    corruption = data_2019.loc[counter:counter+20]["Perceptions of corruption"].mean()
    data["Groups"].append("Group %d"%i)
    data["GDP per capita"].append(gdp)
    data["Social support"].append(social)
    data["Healthy life expectancy"].append(helth)
    data["Freedom to make life choices"].append(freedom)
    data["Generosity"].append(generosity)
    data["Perceptions of corruption"].append(corruption)
    counter+=20
```

```python
df = pd.DataFrame (data, columns = ["Groups", "GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption"])

df
```

```python
plt.figure(figsize=(20,10))
plt.plot(df["GDP per capita"])
plt.plot(df["Social support"])
plt.plot(df["Healthy life expectancy"])
plt.plot(df["Freedom to make life choices"])
plt.plot(df["Generosity"])
plt.plot(df["Perceptions of corruption"])

plt.title("Crop Yields in Kanto")
plt.legend(['GDP per capita', 'Social support', "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption"]);
```

#### Q4: What is the list of countries with highest score in each criteria

```python
max_score = data_2019[data_2019["Score"] == data_2019["Score"].max()]
max_gdp = data_2019[data_2019["GDP per capita"] == data_2019["GDP per capita"].max()]
max_social = data_2019[data_2019["Social support"] == data_2019["Social support"].max()]
max_health = data_2019[data_2019["Healthy life expectancy"] == data_2019["Healthy life expectancy"].max()]
max_freedom = data_2019[data_2019["Freedom to make life choices"] == data_2019["Freedom to make life choices"].max()]
max_gene = data_2019[data_2019["Generosity"] == data_2019["Generosity"].max()]
max_corr = data_2019[data_2019["Perceptions of corruption"] == data_2019["Perceptions of corruption"].max()]
```

```python
countries = {
    "Criteria":["Score", "GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life               choices", "Generosity", "Perceptions of corruption"],

    "Country":[list(max_score.to_dict()["Country or region"].items())[0][1],
        list(max_gdp.to_dict()["Country or region"].items())[0][1],
        list(max_social.to_dict()["Country or region"].items())[0][1],
        list(max_health.to_dict()["Country or region"].items())[0][1],
        list(max_freedom.to_dict()["Country or region"].items())[0][1],
        list(max_gene.to_dict()["Country or region"].items())[0][1],
        list(max_corr.to_dict()["Country or region"].items())[0][1]
    ]}
```

```python
df = pd.DataFrame (countries, columns = ["Criteria", "Country"])
df.style.set_properties(**{'text-align': 'left'}).set_table_styles([ dict(selector='th', props=[('text-align', 'left')] ) ]).hide_index()
```

Here we first get the highest value in each criteria, then extract the country name of the respective value. Then create a dictionary so that it will be converted to a pandas dataframe.
