Here we have declared a python list named "list_1" on line 1 and printed the list and type on the 2nd line.

```python
# Example 2 - working
arr_2 = np.array((11122.55666, 4514.668, 7845), dtype = np.float64)
print(arr_2, type(arr_2))
print(arr_2[1])
```

## Function 2 - concatenate

This function is use to concatenate or join two or more numpy array.

```python
# Example 1 - working

x = np.array((1, 2, 3))
y = np.array((4, 5, 6))
z = np.array((7, 8, 9))

arr_concatinated = np.concatenate((x, y, z))

print(arr_concatinated)
```

Here we have declared 3 numpy array as "x", "y", "z". and passed them into concatenate function and stored the finaly result in arr_concatinated and printed it.

```python
# Example 2 - working
a = np.array([[10.5, 2], [3, 4]])
b = np.array([[5, 6.7], [8, 9.6]])

con_x_axis = np.concatenate((a, b), axis=0)
con_y_axis = np.concatenate((a, b), axis=1)

print(con_x_axis)
print()
print(con_y_axis)
```

In this example we gave one more argument known as "axis". It let us define the way or direction we want to concatinate the multi-dimenssional arrays. In our case we have two 2D arrays and we concatinated them in two axis ie. X-axis in and then Y-axis.

```python
# Example 3 - breaking (to illustrate when it breaks)
m = [[1, 2], [1, 5, 7]]
n = [[1, 9]]

m_n = np.concatenate((m, n))

print(m_n)
```

Before concatinating arrays, Check the dimenssion that they can be concatinated or not.

This is one of this most important function as you may encounter a lot of situation where where you have to join different arryas.

## Function 3 - repeat

This function is used to create an array of a single repeating item.

```python
# Example 1 - working
arr_4 = np.repeat(5.4, 8)

print(arr_4)
```

Here we used repeate function to and passed two argument. First one the item you want to repeate and the second one how many time you want to repeate the item. Here we repeated 5.4, 8 times

Here we have passed an existing array as an argument and repeated the elements of that array n times.

## Function 4 - random

Here actually random is not a funciton it is a class inside NumPy, which holds lot of function to generate random element array.

```python
# Example 1 - working
rand_arr1 = np.random.randint(low = 10, high = 50, size = 10)

print(rand_arr1)
```

Here we have used randint function of random class to generate an array of randon intigers. Here the low argument is the lowest number in range and the high argument is the highest number in range and the size argument tell how many random element is to be stored in the array. 

```python
# Example 2 - working
rand_arr2 = np.random.uniform(low = 10, high = 50, size = (5, 4))

print(rand_arr2)
```

Here we used the uniform function to create random float and we can pass a tuple of numbers to the size argument to create random multi-dimenssional array

```python
# Example 3 - breaking (to illustrate when it breaks)
rand_arr3 = np.random.randint(low = 60, high = 20, size = 10)

print(rand_arr3)
```

Do not use a greater number for low argument than the high argument.

It is a very useful class to use when you are creating a game where you want random actions.

## Function 5 - sort

This function is used to sort a numpy array in may different way and orders.

```python
# Example 1 - working
unsorted_array = np.array([3, 8, 5, 45, 65, 74])

sorted_array = np.sort(unsorted_array)

print(sorted_array)
```

Here we first define a unsorted array with some random numbers. Then pass the unsored array in the NumPy sort function and store the resut in a variable and then print the sorted array.

```python
# Example 2 - working
unsorted_array2 = np.array([[3, 8, 5], [45, 65, 74], [8, 64, 7]])

sorted_array2 = np.sort(unsorted_array2, axis=0, kind='heapsort')

print(sorted_array2)
```

Here we have provided some extra argument in the sort function first is axis, The direction you want to sort. The the "kind" argumnet which have some values like 'quicksort’, ‘mergesort’, ‘heapsort’, ‘stable’. The all have different way of sorting algorithms resulting to different Time Complexity.

```python
# Example 3 - breaking (to illustrate when it breaks)
sorted_array2 = np.sort(unsorted_array2, axis=2, kind='heapsort')

print(sorted_array2)
```

Do not give wrong Axis number.

It is a very useful function, and save you a lot of time while sorting large data sets.
